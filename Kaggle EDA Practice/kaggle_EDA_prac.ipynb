Visualize missing data

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.heatmap(df.isnull(), cbar=False, cmap="YlOrRd")
# df.isnull(): Creates a heatmap showing True (missing) and False (not missing)
# cbar=False: Removes the color bar legend (for cleaner output)
# cmap=YlOrRd: Uses the yellow-orange-red color palette for missing values

plt.title("Missing Values Heatmap")
plt.show()
```

**Decide How to Handle Missing Values**

WORK_CITY = Categorical = âŒ Drop column (too many missing ~2260 / ~5K)

WORK_PROVINCE = Categorical = ðŸ”„ Fill with "Unknown"

WORK_COUNTRY = Categorical = ðŸ”„ Fill with mode (CN)

AGE = Numeric = ðŸ”„ Fill with median age

GENDER = Categorical = ðŸ”„ Fill with mode (Male or Female)

SUM_YR_1, SUM_YR_2 = Numeric = ðŸ”„ Fill with 0 or median (based on use case)


```python
#Drop WORK_CITY since it's too incomplete
df.drop("WORK_CITY", axis=1, inplace=True)

#Fill missing WORK_PROVINCE with "Unknown"
df["WORK_PROVINCE"].fillna("Unknown", inplace=True)

#Fill WORK_COUNTRY with mode
df["WORK_COUNTRY"].fillna(df["WORK_COUNTRY"].mode()[0], inplace=True)

#Fill AGE with median
df["AGE"].fillna(df["AGE"].median(), inplace=True)

#Fill GENDER with mode
df["GENDER"].fillna(df["GENDER"].mode()[0], inplace=True)

#Fill SUM_YR_1 and SUM_YR_2 with 0 (assuming they are unused year columns for some users)
df["SUM_YR_1"].fillna(0, inplace=True)
df["SUM_YR_2"].fillna(0, inplace=True)
```

**Mini Feature Engineering**

Weâ€™ll create user segments based on behavior to support clustering or targeted marketing.

ðŸŽ¯ Goal:

Letâ€™s define customer types using logic/rules on:

Flights

Points

Age

Tiers

***1: Create New Segments***

We'll create 3 example customer segments:

1.Loyal Customers

Earned many points â†’ top 25% in Points_Sum

2.Young Frequent Flyers

Age < 35 and Flight_Count > 75th percentile

3.Infrequent Travelers

Flight_Count < 25th percentile

```python
#Calculate key percentiles
q25_flights = df["FLIGHT_COUNT"].quantile(0.25)
q75_flights = df["FLIGHT_COUNT"].quantile(0.75)
q75_points = df["Points_Sum"].quantile(0.75)
# quantile(0.25) = value below which 25% of data falls â†’ used to define low flight count
# quantile(0.75) = value above which 25% of data falls â†’ used to define high flight count / points
# These help us define thresholds for segment logic like top 25%, bottom 25%, etc.

#Create a new column to store customer segment
df["Customer_Segment"] = "Other"
#We first initialize a new column Customer_Segment with a default label: 'Other'

#Label Loyal Customers
df.loc[df["Points_Sum"] > q75_points, "Customer_Segment"] = "Loyal Cutomer"

#Label Young Frequent Flyers
df.loc[(df["AGE"] < 35) & (df["FLIGHT_COUNT"] > q75_flights), "Customer_Segment"] = "Young Frequent Flyers"
# For customers younger than 35 AND who fly more than the 75th percentile
# We label them 'Young Frequent Flyer'

#Label Infrequent Travelers
df.loc[df["FLIGHT_COUNT"] < q25_flights, "Customer_Segment"] = "Infrequent Travelers"
# For customers whose flight count is below the 25th percentile (i.e., lowest 25% flyers)
# We label them 'Infrequent Traveler'
```

visualize how many users fall into each segment

```python
plt.figure(figsize=(10,6))
sns.countplot(df, x="Customer_Segment", order=df["Customer_Segment"].value_counts().index, palette="Set2")
plt.title("Customer Segment Distribution")
plt.xlabel("Customer Segment")
plt.ylabel("Number of Customers")
plt.grid(axis="y")# Adds horizontal grid lines
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

Customer Segment Distribution â€” Insights Summary

1)Most customers fall under the "Other" category.

->This means they do not meet the criteria to be labeled as loyal, young frequent, or infrequent travelers.

->They may have average or mixed behavior across features like flight count, points, and age.

2)"Loyal Customers" form the second-largest segment.

->These customers earned loyalty points in the top 25% (above the 75th percentile).

->They are likely high-value customers for the business.

3)"Infrequent Travelers" are the third largest group.

->These individuals take fewer flights than 75% of others (below 25th percentile in flight count).

->They may need marketing encouragement or promotions to engage more.

4)"Young Frequent Flyers" are the smallest segment.

->These are customers under 35 years old who take frequent flights (top 25% flight count).

->This niche segment may represent a young, high-value audience worth targeting for loyalty programs or special offers.
